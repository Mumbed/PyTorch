```python
import dgl
import torch
import numpy as np
import networkx as nx
```

```python
g = dgl.graph(([0,0,0,0,0], [1,2,3,4,5]), num_nodes=6)
```

```python
gx = dgl.to_networkx(g)
nx.draw_networkx(gx)
```

```python
g.ndata['value'] = torch.randn(6,3)
g.edata['weight'] = torch.randn(5,4)

g.ndata['value2'] = torch.randn(6,4,2)

print(g.edata['weight'])
print(g.ndata['value2'])
```
