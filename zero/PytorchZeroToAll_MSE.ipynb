```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
W = 1.0
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

def forward(x):
    return x * W
```

```python
def loss(x,y):
    y_pred = forward(x)
    return (y_pred -y) * (y_pred -y)
```

```python
w_list = []
mse_list = []
for W in np.arange(0.0, 4.1, 0.1):
    print("W =", W)
    l_sum = 0
    for x_val, y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum/3)
    w_list.append(W)
    mse_list.append(l_sum/3)
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('W')
plt.show()
```
